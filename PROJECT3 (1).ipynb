```python
"""
================================================================================
PROJECT NAME     : AI Recommendation System
INTERNSHIP       : DecodeLabs Industrial Training Program 2026
PROJECT NUMBER   : Project 3
AUTHOR           : Muhammad Jawad
DESCRIPTION      : Logic-Based Movie Recommendation System
================================================================================
"""

import os
import pandas as pd


class RecommendationSystem:
    """
    A logic-based movie recommendation system.
    """

    def __init__(self, file_path: str):
        """
        Load the CSV dataset and prepare it for recommendations.
        """
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"Dataset not found: {file_path}")

        self.movies = pd.read_csv(file_path)

        self.movies.columns = self.movies.columns.str.strip()
        self.movies = self.movies.fillna("")

        required_columns = {"Movie", "Genre", "Language", "Rating", "Year"}
        missing_columns = required_columns - set(self.movies.columns)

        if missing_columns:
            raise ValueError(
                f"Missing required columns in dataset: {', '.join(sorted(missing_columns))}"
            )

        self.movies["Rating"] = pd.to_numeric(self.movies["Rating"], errors="coerce")
        self.movies["Year"] = pd.to_numeric(self.movies["Year"], errors="coerce")
        self.movies = self.movies.dropna(subset=["Rating", "Year"])
        self.movies["Year"] = self.movies["Year"].astype(int)

        print("\nDataset Preview")
        print("-" * 30)
        print(self.movies.head())

    def display_genres(self):
        """
        Display all available genres.
        """
        genres = sorted(self.movies["Genre"].dropna().astype(str).unique())

        print("\nAvailable Genres")
        print("-" * 30)
        for genre in genres:
            print(f"• {genre}")

    def get_user_input(self):
        """
        Get user preferences.
        """
        print("\nEnter Your Preferences")
        print("-" * 30)

        genre = input("Genre: ").strip().lower()
        language = input("Language: ").strip().lower()

        while True:
            try:
                minimum_rating = float(input("Minimum Rating (0-10): "))
                if 0 <= minimum_rating <= 10:
                    break
                print("Rating must be between 0 and 10.")
            except ValueError:
                print("Invalid rating. Please enter a numeric value.")

        return genre, language, minimum_rating

    def calculate_similarity(self, movie, genre, language, minimum_rating):
        """
        Calculate a similarity score for one movie.
        """
        score = 0

        movie_genre = str(movie["Genre"]).strip().lower()
        movie_language = str(movie["Language"]).strip().lower()
        movie_rating = float(movie["Rating"])
        movie_year = int(movie["Year"])

        if movie_genre == genre:
            score += 50

        if movie_language == language:
            score += 20

        if movie_rating >= minimum_rating:
            score += 20

        if movie_year >= 2020:
            score += 10
        elif movie_year >= 2015:
            score += 5

        return score

    def recommend(self, genre, language, minimum_rating):
        """
        Generate top movie recommendations.
        """
        recommendations = self.movies.copy()

        recommendations["Score"] = recommendations.apply(
            lambda movie: self.calculate_similarity(
                movie,
                genre,
                language,
                minimum_rating,
            ),
            axis=1,
        )

        recommendations = recommendations.sort_values(
            by=["Score", "Rating", "Year"],
            ascending=[False, False, False],
        )

        return recommendations.head(5)

    @staticmethod
    def display_results(results):
        """
        Display results.
        """
        print("\n")
        print("=" * 70)
        print("TOP RECOMMENDATIONS")
        print("=" * 70)

        if results.empty:
            print("No recommendations found.")
            return

        for i, (_, movie) in enumerate(results.iterrows(), start=1):
            print(f"\n{i}. {movie['Movie']}")
            print("-" * 40)
            print(f"Genre      : {movie['Genre']}")
            print(f"Language   : {movie['Language']}")
            print(f"Rating     : {movie['Rating']}")
            print(f"Year       : {movie['Year']}")
            print(f"Score      : {movie['Score']}")

    @staticmethod
    def goodbye():
        """
        Print goodbye message.
        """
        print("\nThank you for using AI Recommendation System!")
        print("Good Luck with your DecodeLabs Internship.")


def main():
    """
    Program entry point.
    """
    print("=" * 70)
    print("AI RECOMMENDATION SYSTEM".center(70))
    print("=" * 70)

    file_path = r"C:\Users\rcm\Downloads\movies.csv"
    system = RecommendationSystem(file_path)

    while True:
        system.display_genres()
        genre, language, rating = system.get_user_input()

        recommendations = system.recommend(genre, language, rating)
        system.display_results(recommendations)

        again = input("\nDo you want another recommendation? (Y/N): ").strip().lower()
        if again != "y":
            system.goodbye()
            break


if __name__ == "__main__":
    main()
```
